label-encoding
- convert labels into numeric form

```python
# import libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

1 - label encoding of breast cancer dataset

```python
# load dataset from csv file to dataframe
cancer_data = pd.read_csv("/content/data.csv")
```

```python
cancer_data.head()
```

```python
# finding the count of different labels
cancer_data["diagnosis"].value_counts()
```

```python
# load the label encoder function
label_encode = LabelEncoder()
```

```python
# encode all the labels to either 0 or 1 and then store it into separate variables
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# append the labels to the dataframe
cancer_data["target"] = labels
```

```python
cancer_data.head()
```

0 -- benign,
1 -- malignant

```python
cancer_data["target"].value_counts()
```

2 - label encoding of iris dataset

```python
# load dataset from csv file to dataframe
iris_data = pd.read_csv("/content/iris_data.csv")
```

```python
iris_data.head()
```

```python
# finding the count of different labels
iris_data["Species"].value_counts()
```

```python
# load the label encoder function
label_encode_1 = LabelEncoder()
```

```python
# encode all the labels to either 0, 1, 2 and then store it into separate variables
iris_labels = label_encode_1.fit_transform(iris_data.Species)
```

```python
# append the labels to the dataframe
iris_data["target"] = iris_labels
```

```python
iris_data.head()
```

iris-setosa -- 0,
iris-versicolor -- 1,
iris-virginica -- 2

```python
iris_data["target"].value_counts()
```
