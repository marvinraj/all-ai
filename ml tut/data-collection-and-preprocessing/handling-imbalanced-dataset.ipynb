imbalanced dataset:
- a dataset with an unequal distribution

```python
# import libraries
import numpy as np
import pandas as pd
```

```python
# load dataset into dataframe
credit_data = pd.read_csv("/content/credit_data.csv")
```

```python
credit_data.head()
```

```python
credit_data.tail()
```

```python
# distribution of the two classes
credit_data["Class"].value_counts()
```

0 --> legit transactions
1 --> fraudulant transactions

```python
# separate the legit and fraudulent transactions
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

undersampling

number of fraudulant transactions --> 356

```python
# build a sample dataset containing similar distribution of legit and fraudulant
legit_sample = legit.sample(n=356)
```

```python
print(legit_sample.shape)
```

```python
# concatenate the 2 dataframes
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```
