**2 - initial placeholders in numpy array**

```python
# create a numpy array of zeros
x = np.zeros((4,5))
print(x)
```

```python
# create a numpy array of zeros
y = np.ones((3,4))
print(y)
```

```python
# create a numpy array of a specific value
z = np.full((3,3),5)
print(z)
```

```python
# create an identity matrix
a = np.eye(4)
print(a)
```

```python
# create numpy array with random values (between 0 and 1)
b = np.random.random((3,4))
print(b)
```

```python
# create numpy array with random values (integer values)
c = np.random.randint(10,100,(3,6))
print(c)
```

```python
# create numpy array with evenly spaced values (specifying the number of values needed)
d = np.linspace(10,30,5)
print(d)
```

```python
# create numpy array with evenly spaced values (specifying the step)
e = np.arange(10,30,5)
print(e)
```

```python
# convert list into a numpy array
list2 = [1,2,3,4,5]

np_arr = np.asarray(list2)
print(np_arr)
```

**4 - mathematical operations on numpy arrays**

```python
# method 1 - add, subtract, multiply and divide
a = np.random.randint(0,10,(3,3))
b = np.random.randint(10,20,(3,3))

print(a)
print(b)
```

```python
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
# method 1 - add, subtract, multiply and divide
a = np.random.randint(0,10,(3,3))
b = np.random.randint(10,20,(3,3))

print(a)
print(b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

**5 - array manipulation**

```python
# method 1 - transpose
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
trans = np.transpose(array)
print(trans)
print(trans.shape)
```

```python
# method 2 - transpose
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
trans2 = array.T
print(trans2)
print(trans2.shape)
```

```python
# reshape a numpy array
a = np.random.randint(0,18,(2,3))
print(a)
print(a.shape)
```

```python
b = a.reshape(3,2)
print(b)
print(b.shape)
```
